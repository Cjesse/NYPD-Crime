```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

raw_data = 'partII/deliverable/output/KY_CD_BORO_NM_count.out'
df = pd.read_csv(raw_data, header=None)
df = df.rename(columns={0:'KY_CD', 1:'BORO_NM', 2:'COUNT'})
df['KY_CD'] = df['KY_CD'].str.strip('(')
df['BORO_NM'] = df['BORO_NM'].str.strip(')')
df['COUNT'] = df['COUNT'].str.strip(')').astype(np.int64)
# print(df)
# df.head()
plt.figure(figsize=(30,4))
x = df['KY_CD']
y = df['COUNT']
plt.bar(x, y)
plt.gcf().autofmt_xdate()
plt.show()
```

```python
import pandas as pd
import numpy as np
import scipy as sp
import plotly as py
from plotly.graph_objs import *
import plotly.figure_factory as ff
py.offline.init_notebook_mode(connected=True)
# print('Python version ' + sys.version)
# print('Pandas version ' + pd.__version__)
# print('Matplotlib version ' + matplotlib.__version__)

raw_data = 'partII/deliverable/output/KY_CD_BORO_NM_count.out'
df = pd.read_csv(raw_data, header=None)
df = df.rename(columns={0:'KY_CD', 1:'BORO_NM', 2:'COUNT'})

df['KY_CD'] = df['KY_CD'].str.strip('(')
df['BORO_NM'] = df['BORO_NM'].str.strip(')').str.strip(' ').str.strip("'")
df['COUNT'] = df['COUNT'].str.strip(')')
df.head()
df.shape

booleans = []
for item in df.BORO_NM:
    if item == '':
        booleans.append(False)
    else:
        booleans.append(True)
        
is_long = pd.Series(booleans)
is_long.head()
df = df[is_long]
table = df.pivot(index='KY_CD', columns = 'BORO_NM', values = 'COUNT')
table.to_csv('test.csv')
table.columns
```
