```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

### total crimes each year from 2006 - 2016

```python
filepath = '/Users/suyangjiang/Desktop/big data analytics/project/visualization/data/'
```

#### crimes for each year

```python
df = pd.read_csv(filepath + 'year_count.out', header=None)
df = df.rename(columns={0:'Year', 1:'Crime Total'})
df['Year'] = df['Year'].str.strip('(').astype(np.int64)
df['Crime Total'] = df['Crime Total'].str.strip(')').astype(np.int64)
x = df['Year']
y = df['Crime Total']
plt.figure(figsize=(10,4))
plt.plot(x, y)
plt.xticks(np.arange(2006, 2017, 1))
plt.xlabel('year')
plt.ylabel('total crime')
plt.title('total crimes each year')
plt.show()
```

```python
df = pd.read_csv(filepath + 'end_year_count.out', header=None)
df = df.rename(columns={0:'Year', 1:'Crime Total'})
df['Year'] = df['Year'].str.strip('(').astype(np.int64)
df['Crime Total'] = df['Crime Total'].str.strip(')').astype(np.int64)
x = df['Year']
y = df['Crime Total']
plt.figure(figsize=(10,4))
plt.plot(x, y)
plt.xticks(np.arange(2006, 2017, 1))
plt.xlabel('END-year')
plt.ylabel('total crime')
plt.title('total crimes END each year')
plt.show()
```
